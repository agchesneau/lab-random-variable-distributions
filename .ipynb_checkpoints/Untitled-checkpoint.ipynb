```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
customer_df = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
customer_df.head()
```

### Instructions

Get the numerical variables from our dataset.

Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
numerical = customer_df.select_dtypes(np.number)
```

```python
for data in numerical.columns :
    sns.displot(numerical[data], bins=30)
    plt.show()
```

```python
# customer lifetime value, monthly preÃ¹iuÃ¹ auto, nb of open complaints, number of policies look exponentianlly distributed
# none look like a normal distribution
```

```python
def log_transfom(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0
    
numerical_log = pd.DataFrame(data = None)    
for col in numerical :
    numerical_log[col] = numerical[col].apply(lambda x : log_transfom(x))
```

```python
numerical_log.head()
```

```python
for data in numerical_log.columns :
    sns.displot(numerical_log[data], bins=50)
    plt.show()
```

```python
#customer lifetime value and total claim amount look more like a normal distribution
```

```python
from sklearn.preprocessing import MinMaxScaler
normalized = pd.DataFrame(MinMaxScaler().fit_transform(numerical), columns=numerical.columns)

for data in normalized.columns :
    sns.displot(normalized[data], bins=50)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
standardized = pd.DataFrame(StandardScaler().fit_transform(numerical), columns=numerical.columns)

for data in standardized.columns :
    sns.displot(standardized[data], bins=50)
    plt.show()
```

```python
# these transformations don't change much at all in the ditribution
```
